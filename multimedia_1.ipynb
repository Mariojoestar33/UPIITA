```python
#Handler del archivo: file
#Leer
#Ubicarme en alguna parte del archivo

file = open("./images/example001.bmp","rb")

firm = file.read(2)
print(firm) #Verifia la firma para saber si es de un archivo BMP
file.seek(84,0) #Va a la posicion 84
pixel_first = file.read(3) #Toma los 3 numeros siguientes
print(pixel_first)
print(int.from_bytes(pixel_first,byteorder='big')) #Representa en Entero
print(int.from_bytes(pixel_first,byteorder='little')) #Voltea el orden de los bytes

file.seek(54,0)
rojo = 0
verde = 0
azul = 0
blanco = 0
negro = 0
no_pix = 0
while(True):
    pixel_data = file.read(3)
    if(len(pixel_data) > 0):
        #print(pixel_data)
        print(int.from_bytes(pixel_data,byteorder='little'))
        no_pix += 1
        if(int.from_bytes(pixel_data,byteorder='little') == 16711680):
          rojo +=1;
        if(int.from_bytes(pixel_data,byteorder='little') == 255):
          azul +=1;
        if(int.from_bytes(pixel_data,byteorder='little') == 65280):
          verde +=1;
        if(int.from_bytes(pixel_data,byteorder='little') == 16777215):
          blanco +=1;
        if(int.from_bytes(pixel_data,byteorder='little') == 0):
          negro +=1;
    else:
        break
print('No Pixels: '+str(no_pix))
totalr = rojo/4
totala = azul/4
totalv = verde/4
totalb = blanco/4
totaln = negro/4
print('Numero total de elementos rojos = ' +str(rojo))
print('Numero total de elementos azul = ' +str(azul))
print('Numero total de elementos verde = ' +str(verde))
print('Numero total de elementos blanco = ' +str(blanco))
print('Numero total de elementos negro = ' +str(negro))
file.close()
```
